```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker
import plotly.express as px


dat = pd.read_csv('./historical_index.csv',
                   # usecols=['DEST_STATE_NM' ,'DAY_OF_MONTH','ARR_DELAY','AIR_TIME','DISTANCE']
                   )
#dat.drop(['HDI Rank'], axis=1)
#dat=dat.dropna(axis=0)
#dat[dat["Human Development Index (1990)"].str.contains( ".." )==False]
#dat = dat[dat["Human Development Index (1990)"].str.contains("..") == False]
dat = dat[dat['Human Development Index (1990)'] != '..']
#dat.dtypes
#dat = dat[dat["Human Development Index (1990)"].str.contains("|..") == False]
dat['Human Development Index (1990)'] = dat['Human Development Index (1990)'].astype(float)
dat['Human Development Index (2000)'] = dat['Human Development Index (2000)'].astype(float)
dat.head(10)
dat.dtypes
#HDI increasing rate from 1990 to 2014
increase = dat['Human Development Index (2014)'] - dat['Human Development Index (1990)']
IR = increase/dat['Human Development Index (1990)'] * 100
IR = IR.round(decimals = 2)
dat.loc[:, "Increasing Rate Percentage"] = IR
#dat.head(50)

#dat = dat.sort_values('Increasing Rate)', ascending=False)
#sss=dat['Increasing Rate'].iloc[0:5]
#sss.head(6)
Iranking = dat['Increasing Rate Percentage'].rank(method='average', ascending=False)
Iranking = Iranking.astype(int)
Iranking = Iranking.astype(str)
dat.loc[:, "Increasing Rate Ranking"] = Iranking
dat = dat.sort_values('Increasing Rate Percentage', ascending=False)
dat.dtypes

#fig1 = px.histogram(dat,x = 'Human Development Index (1990)',y='Increasing Rate', color='Country',
                   #hover_name="Country",animation_frame="Country").update_xaxes(categoryorder="total ascending")

fig1 = px.histogram(dat.iloc[0:50], x = 'Increasing Rate Ranking',y='Increasing Rate Percentage', color='Country',
                    hover_name="Country",

                   width=1200, height=800).update_xaxes(categoryorder="total ascending")

fig1.update(layout_yaxis_range = [0,110])
fig1.update_layout(barmode='stack',xaxis={'categoryorder':'total descending'},margin=dict(l=55, r=20, t=40, b=5,pad=0),
                   title_text='The Increasing Rate of HDI from 1990 to 2014', # title of plot
                    xaxis_title_text='Increasing Rate Ranking', # xaxis label
                    yaxis_title_text='Increasing Rate Percent %', # yaxis label
                    bargap=0.3 # gap between bars of adjacent location coordinates
                  )

fig1.show()
#based on the graph above, we could see the most increasing countries from 1990 to 2014 are Rawada,
#Mozambique, Senegal, Mali, Niger, etc.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import random
import csv
import math
import matplotlib.ticker as mticker
import plotly.express as px
import plotly
import plotly.graph_objects as go

import plotly.io as pio

pio.renderers.default = 'colab'  # Set the correct renderer for Colab

datas = pd.read_csv('./historical_index.csv', usecols=['Country', 'Human Development Index (1990)', 'Human Development Index (2014)'])
# datas = datas.dropna()

datas['Human Development Index (1990)'] = pd.to_numeric(datas['Human Development Index (1990)'], errors='coerce')
datas['Human Development Index (2014)'] = pd.to_numeric(datas['Human Development Index (2014)'], errors='coerce')

datas.drop(datas.index[0:188], inplace=True)
print(datas)

state = ['Arab States', 'East Asia and the Pacific', 'Europe and Central Asia', 'Latin America and the Caribbean', 'South Asia', 'Sub-Saharan Africa', 'World']

r = [0.553, 0.516, 0.651, 0.625, 0.437, 0.400, 0.597]
r1 = [0.686, 0.710, 0.748, 0.748, 0.607, 0.518, 0.711]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=r,
      theta=state,
      fill='tonext',
      name='HDI in 1990'
))

fig.add_trace(go.Scatterpolar(
      r=r1,
      theta=state,
      fill='tonext',
      name='HDI in 2014'
))

fig.update_layout(
  title="Comparison of HDI in different regions in 1990 and 2014",
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 1]
    )),
  showlegend=True
)

fig.show()
```

```python
datas = pd.read_csv('./human_development.csv',usecols=['Country' ,'Human Development Index (HDI)','Life Expectancy at Birth'])

highdatas = datas
highdatas = highdatas.dropna()
HDI = highdatas['Human Development Index (HDI)'].to_list()

for i in range(0, len(HDI)):
    if(HDI[i]<0.6):
        highdatas.drop([i],axis=0,inplace=True)
highdatas = highdatas.reset_index(drop=True)

lift = highdatas['Life Expectancy at Birth'].to_list()
highlight = 0
shortlife = 0

for i in range(0, len(lift)):
    if(lift[i]>70):
        highlight = highlight+1
    else:
        shortlife = shortlife+1

year = pd.DataFrame({
    "name":["Normal life","Longer life"],
    "number":[shortlife,highlight]})


fig = px.pie(year,names="name",values="number",title='High HDI longevity ratio', color="name",
             color_discrete_map={
                 'Normal life':'#faff72',
                 'Longer life':'#bce672'
                 }
            )

fig.show()
```

```python
lowdatas = datas
lowdatas = lowdatas.dropna()
HDI1 = lowdatas['Human Development Index (HDI)'].to_list()

for i in range(0, len(HDI1)):
    if(HDI1[i]>0.6):
        lowdatas.drop([i],axis=0,inplace=True)
lowdatas = lowdatas.reset_index(drop=True)

lift1 = lowdatas['Life Expectancy at Birth'].to_list()
highlight1 = 0
shortlife1 = 0

for i in range(0, len(lift1)):
    if(lift1[i]>70):
        highlight1 = highlight1+1
    else:
        shortlife1 = shortlife1+1

year1 = pd.DataFrame({
    "name":["Longer life","Normal life"],
    "number":[highlight1,shortlife1]})
fig = px.pie(year1,names="name",values="number",title='Low HDI longevity ratio', color="name",
             color_discrete_map={
                 'Normal life':'#faff72',
                 'Longer life':'#bce672'
                 }
            )
fig.show()
```
